```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import spectrogram
import os

# ==== SETTINGS ====
csv_path = "both small.csv"     # path to your CSV
fs = 100                  # sampling frequency in Hz (CHANGE THIS)
output_folder = "spectrograms"

# ==================

os.makedirs(output_folder, exist_ok=True)

# Load CSV
df = pd.read_csv(csv_path)

# Columns to process
columns = ["wx", "wy", "wz", "ax", "ay", "az"]

for col in columns:
    if col not in df.columns:
        print(f"Column {col} not found. Skipping.")
        continue

    signal = df[col].values

    # Compute spectrogram
    f, t, Sxx = spectrogram(signal, fs=fs)

    # Plot
    plt.figure(figsize=(8,5))
    plt.pcolormesh(t, f, Sxx, shading='gouraud')
    plt.ylabel("Frequency (Hz)")
    plt.xlabel("Time (sec)")
    plt.title(f"Spectrogram - {col}")
    plt.colorbar(label="Intensity")

    # Save image
    save_path = os.path.join(output_folder, f"{col}_spectrogram.png")
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.close()

    print(f"Saved {save_path}")
```
